```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Veri Seti
data = pd.read_csv("Mall_Customers.csv")
```

```python
# Özniteliklerin seçimi (Özellikler: 'Age', 'Annual Income (k$)', 'Spending Score (1-100)')
x = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x_scaled)
data['cluster'] = kmeans.labels_
```

```python
#Kümeleri görselleştirme (3D örneği)
fig = plt.figure(figsize=(12, 20))
ax = fig.add_subplot(111, projection='3d')
# Verileri ve kümeleri 3D olarak çizme
ax.scatter(x_scaled[:, 0], x_scaled[:, 1], x_scaled[:, 2], c=kmeans.labels_, cmap='plasma', s=150)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], c='red', marker='*', s=200, label='Küme Merkezleri')
plt.tight_layout()  # Grafik düzenini sıkıştırmadan genişlet
# Grafik başlıkları ve etiketler
title = "K-Means Kümeleme (3D)"
ax.set_title(title, color = "red",fontsize=20)
ax.set_xlabel("Age", color = "red",fontsize=20)
ax.set_ylabel("Annual Income (k$)", color="red",fontsize=20)
ax.set_zlabel("Spending Score (1-100)",color = "red",fontsize=20)
plt.show()
```
